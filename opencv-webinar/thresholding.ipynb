## Global Thresholds

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('grad.jpg',0)
_,t1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
_,t2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
_,t3 = cv.threshold(img,127,255,cv.THRESH_TRUNC)
_,t4 = cv.threshold(img,127,255,cv.THRESH_TOZERO)
_,t5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)

titles = ['Original','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, t1, t2, t3, t4, t5]
for i in range(6):
    plt.subplot(2,3,i+1), plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```

## Adaptive Thresholds

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('sudoku.jpg',0)
img = cv.medianBlur(img,1)

_,th1 = cv.threshold(img,150,255,cv.THRESH_BINARY)
th2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,\
            cv.THRESH_BINARY,51,2)
th3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY,51,2)

titles = ['Original','BINARY','Mean','Gaussian']
images = [img, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
